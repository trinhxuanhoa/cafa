```python
import pandas as pd
from Bio import SeqIO
import matplotlib.pyplot as plt

# 1. Đọc dữ liệu chuỗi Protein (FASTA)
# File này chứa: ID protein và chuỗi axit amin (VD: MKTLL...)
fasta_file = "../data/raw/train_sequences.fasta"
sequences = []
for record in SeqIO.parse(fasta_file, "fasta"):
    sequences.append({"EntryID": record.id, "Sequence": str(record.seq), "Length": len(record.seq)})

df_seq = pd.DataFrame(sequences)
print(f"Tổng số protein: {len(df_seq)}")
print(df_seq.head())

# 2. Đọc dữ liệu nhãn (Labels) - Chức năng của protein
# File này chứa: ID protein và mã GO (Gene Ontology) tương ứng
terms_file = "../data/raw/train_terms.tsv"
df_terms = pd.read_csv(terms_file, sep="\t")
print(f"\nTổng số dòng nhãn: {len(df_terms)}")
print(df_terms.head())

# 3. Vẽ biểu đồ độ dài Protein (Xem protein dài ngắn thế nào)
plt.figure(figsize=(10, 5))
plt.hist(df_seq['Length'], bins=100, range=(0, 2000), color='skyblue', edgecolor='black')
plt.title("Phân phối độ dài Protein")
plt.xlabel("Số lượng axit amin")
plt.ylabel("Số lượng protein")
plt.show()
```
